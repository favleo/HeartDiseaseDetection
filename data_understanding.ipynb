```python
# Stage 4 - Check for potential leakage
## Define suspect features (all columns labeled as 'feature')
features = [
    'Sex', 'GeneralHealth', 'PhysicalHealthDays', 'MentalHealthDays',
    'PhysicalActivities', 'SleepHours', 'RemovedTeeth', 'HadAngina',
    'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD',
    'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
    'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'SmokerStatus', 'ECigaretteUsage',
    'RaceEthnicityCategory', 'AgeCategory', 'HeightInMeters',
    'WeightInKilograms', 'BMI', 'AlcoholDrinkers'
]

df_numeric = df[features + ['HadHeartAttack']].apply(lambda x: pd.factorize(x)[0])

correlations = df_numeric.corr()
print("Correlation matrix:\n", correlations)

print("\n" + "="*60)
print("Leakage Detection - Correlation with Target (HadHeartAttack)")
print("="*60)

for col in features:
    leak_target_corr = correlations.loc[col, 'HadHeartAttack']
    risk_level = "HIGH RISK" if abs(leak_target_corr) > 0.5 else "MODERATE" if abs(leak_target_corr) > 0.3 else "LOW"
    print(f"Correlation of '{col:30s}' with HadHeartAttack: {leak_target_corr:7.4f} [{risk_level}]")
```
